# Convert est_price (object type) to a numeric value

```python
Coffee_df.est_price.unique()
```

Lots of variance in how prices are recorded.  Wide arrange of currencies and units of measurements.  Some numbers are stored as the written word which makes conversion to int and regex matching difficult.

The most common units of measurements are ounces and grams.
Rare occurences (pounds)

The most common currencies are USD, NT, and HKD.
Rare occurences (yen, pound, euros, CAD)

Plan: Use regex to extract the numeric portions, currency, and the unit (ounce/gram).  See what percentage of data is captured.

```python
# remove unwanted characters
Coffee_df['est_price'] = Coffee_df['est_price'].map(lambda x: str(x).replace(',',''))
```

```python
# extract the numeric portion of the price
Coffee_df['value'] = Coffee_df['est_price'].str.extract('([0-9]*\.*[0-9]+/[0-9]*\.*[0-9]+)')
Coffee_df.value.unique()
```

```python
Coffee_df.value.count()
```

```python
# extract the unit of measurement
Coffee_df['unit'] = Coffee_df['est_price'].str.extract('(ounces|grams|pounds)')
Coffee_df.unit.unique()
```

```python
Coffee_df['unit'].count()
```

```python
# extract the currency
Coffee_df['currency'] = Coffee_df['est_price'].str.extract('(NT|HKD|CAD|KRW|AUD|AED|IDR|€|£|¥|RM|[eE]uro)')
Coffee_df.currency.unique()
```

```python
Coffee_df.currency.count()
```

```python
# If unknown currency, assume USD
Coffee_df.currency.fillna('$', inplace=True)
Coffee_df.currency.unique()
```

```python
# remove rows that could not be converted
Coffee_df = Coffee_df.dropna(subset=['currency', 'value', 'unit'])
Coffee_df.shape
```

```python
Coffee_df[['currency', 'value', 'unit']]
```

```python
import sympy

#Coffee_df[['value']].apply(lambda x: sympy.sympify(x))
#Coffee_df[['value']].eval('value', target=Coffee_df)



def convert(s):
    try:
        return float(s)
    except ValueError:
        num, denom = s.split('/')
        return float(num) / float(denom)

Coffee_df['value'] = Coffee_df['value'].apply(lambda x: convert(x))
Coffee_df.astype({'value': 'float'}).dtypes
Coffee_df[['value']]
```

# Sentiment Analysis

First clean the texts by removing numbers and special symbols.  Then analyze using VADER sentiment analysis, a lexicon-based analyzer.

Vader was pretrained using social media.  This allows it to be applied directly to the coffee reviews without needing to be trained.  Pretrained models are good for small datasets because the data does not need to be split into training and test sets.

Lexicon-based analyers remove the need for a greater context because it breaks down texts into individual tokens instead of viewing the greater context.  making them a general solution for sentiment analysis without worry of being overly fit.

```python
import nltk
import re

# remove special characters
def clean(text):
    text = re.sub('[^A-Za-z]+', ' ', str(text))
    return text

# place cleaned descriptions into new columns with identifiable names
Coffee_df['blind_assessment'] = Coffee_df['desc_1'].apply(clean)
Coffee_df['notes'] = Coffee_df['desc_2'].apply(clean)
Coffee_df['bottom_line'] = Coffee_df['desc_3'].apply(clean)
Coffee_df[['blind_assessment', 'notes', 'bottom_line']].head()
```

```python
# https://stackoverflow.com/questions/40325980/how-is-the-vader-compound-polarity-score-calculated-in-python-nltk
# polarity_scores output into 4 classes of sentiments (neg, neu, pos, compound (aggregated score))

analyzer = SentimentIntensityAnalyzer()

def analyze_text(text):
    sentiment = analyzer.polarity_scores(str(text))
    return sentiment['compound']


Coffee_df['blind_assessment_sentiment'] = Coffee_df['blind_assessment'].apply(analyze_text)
Coffee_df['notes_sentiment'] = Coffee_df['notes'].apply(analyze_text)
Coffee_df['bottom_line_sentiment'] = Coffee_df['bottom_line'].apply(analyze_text)

Coffee_df[['blind_assessment', 'blind_assessment_sentiment']]
```
