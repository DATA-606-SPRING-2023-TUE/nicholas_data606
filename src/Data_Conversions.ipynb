# Import Libraries


```python
import pandas as pd
```

# Read the CSV File

```python
Coffee_df = pd.read_csv("https://raw.githubusercontent.com/DATA-606-SPRING-2023-TUE/nicholas_data606/main/data/coffee_df_with_type_and_region.csv")
Coffee_df = Coffee_df.drop(columns=['with_milk'])
Coffee_df.head()
```

# DATA CONVERSIONS


Convert review_date Strings to datetime

```python
Coffee_df.review_date.unique
```

```python
Coffee_df['review_date'] = pd.to_datetime(Coffee_df['review_date'], format='%B %Y')
Coffee_df['review_date']
```

Convert Price per ounce (object type) to a numeric value

```python
Coffee_df['est_price']
```

```python
#Coffee_df['value'] = Coffee_df['est_price'].str.extract("(\d+\.\d+\/\d+\.\d+\)")
Coffee_df['value'] = Coffee_df['est_price'].str.extract("([0-9]*.[0-9]*/[0-9]*.[0-9]*)")
Coffee_df['value']
```

```python
Coffee_df[['before']] = Coffee_df["est_price"].apply(lambda x: pd.Series(str(x).rsplit(r"([0-9]*.[0-9]*/[0-9]*.[0-9]*)")))
Coffee_df.before.unique()
```

```python
# Split single column into two columns use apply()
Coffee_df[['value', 'unit']] = Coffee_df["est_price"].apply(lambda x: pd.Series(str(x).rsplit(" ", 1)))
Coffee_df[['currency', 'value']] = Coffee_df["value"].apply(lambda x: pd.Series(str(x).rsplit("$", 1)))
Coffee_df[['currency', 'value', 'unit']]
```
