# Check the existence of Null Values

```python
Coffee_df.isnull().sum()
```

```python
# Drop the slug and all_text columns because they provide no data (one is a text link and the other is the full text of the review in an unreadable format)
# Drop the with_milk feature because of the high percentage of null values

Coffee_df = Coffee_df.drop(columns=['slug', 'all_text', 'with_milk'])
Coffee_df.isnull().sum()
```

```python
# drop rows with null values in key columns.  acidity, aroma, roast and description were to be used as features.  Price can be used to gauge demand 
Coffee_df = Coffee_df[Coffee_df['acid'].notna()]
Coffee_df = Coffee_df[Coffee_df['est_price'].notna()]
Coffee_df = Coffee_df[Coffee_df['aroma'].notna()]
Coffee_df = Coffee_df[Coffee_df['desc_3'].notna()]
Coffee_df = Coffee_df[Coffee_df['roast'].notna()]
Coffee_df.isnull().sum()
```
